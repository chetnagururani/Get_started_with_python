# Introduction to Python

1. <b>Why do we need a programming language?
- A computer is a versatile machine that can perform millions of instructions , but can neither think nor make decisions on its own . It requires some instruction, thus a program is a set of instructions that guides one to solve a particular task.

 <b> Python 
- Python is a high-level, general-purpose, and interpreted programming language known for its simplicity

- "High-level": refers to a language that is closer to human language, High-level languages manage complex details like memory allocation and CPU instructions automatically, so the programmer doesn't have to handle them directly.
- Interpreted programming language: Python reads your code line by line. This process is called interpreting. Instead of converting your entire program into machine language all at once (like a compiled language), Python reads your code line by line.
- General purpose: A general-purpose programming language is a language that can be used to write software for a wide variety of tasks and applications, rather than being specialized for a specific type of problem or domain.

<b> Python history

- Python was created in the late 1980s by Guido van Rossum
  
- Released publicly in 1991, Python was designed as a successor to the ABC language, aiming to be easy to learn while being more flexible. -
- Python 1.0 was officially released in 1994, introducing core features like exception handling and functional programming tools.
- In 2000, Python 2.0 added important enhancements like garbage collection and list comprehensions, making the language more robust. However,   Python 2 had limitations, leading to the release of Python 3.0 in 2008
- Python 3 was not backward-compatible with Python 2, and in January 2020, Python 2 officially reached the end of its life. Since then,
- Python 3 has continued to evolve, becoming one of the most popular programming languages due to its simplicity, versatility, and extensive libraries for tasks ranging from web development to data science.
